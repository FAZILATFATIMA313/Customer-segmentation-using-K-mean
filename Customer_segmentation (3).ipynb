#Elbow Method for Optimal K in K-Means Clustering
This section uses the Elbow Method to find the optimal number of clusters (k) for K-Means clustering by evaluating how the Within-Cluster Sum of Squares (WCSS) changes with different k values.

* **Counting WCSS:**
We calculate the WCSS for k values from 1 to 10. Lower WCSS means better clustering, as the data points are closer to their centroids.

* **Plotting the Elbow Curve:**
We plot the WCSS against k values and add red markers to highlight the data points. The x-axis represents the number of clusters, and the y-axis represents WCSS.

* **Interpreting the Elbow:**
The optimal k corresponds to the "elbow" point where the WCSS starts to decrease at a slower rate. This point indicates the best balance between model complexity and accuracy.

```python
# Cell 7: Elbow Method for Optimal K in K-Means Clustering

# Initialize an empty list to store the WCSS (Within-Cluster Sum of Squares) values
wcss = []

# Loop to calculate WCSS for different values of 'k' (number of clusters)
for i in range(1, 11):
    km = KMeans(n_clusters=i)  # Create a KMeans model with i clusters
    km.fit(X)  # Fit the model to the data X
    wcss.append(km.inertia_)  # Append the WCSS (inertia) value for this 'k'

# The Elbow Curve - to visualize the relationship between the number of clusters and WCSS
plt.figure(figsize=(12, 6))  # Set the figure size for better readability
plt.plot(range(1, 11), wcss)  # Plot WCSS values for k values from 1 to 10
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="8")  # Add red markers to the plot
plt.xlabel("K Value")  # Label the x-axis as the number of clusters (k)
plt.xticks(np.arange(1, 11, 1))  # Set x-axis tick marks to range from 1 to 10 with step size 1
plt.ylabel("WCSS")  # Label the y-axis as the WCSS value
plt.title("Elbow Method for Optimal K")  # Add a title to the plot
plt.show()  # Display the plot
```

#Customer Segmentation using K-Means Clustering
In this section, we perform customer segmentation based on two features: Annual Income (k$) and Spending Score (1-100). We use K-Means clustering to group customers into distinct segments and analyze their spending behavior.

* **Data Normalization:**
The first step is normalizing the data using MinMaxScaler. We scale the Annual Income and Spending Score to a range of [0, 1] to ensure that both features contribute equally to the clustering process.

* **Scatter Plot:**
After normalization, we plot the Annual Income vs Spending Score to visualize how the data is distributed before applying the clustering algorithm.

* **K-Means Clustering:**
We apply the K-Means clustering algorithm to group the customers into 5 clusters. The algorithm assigns each data point to one of the clusters based on its proximity to the cluster centroids.

* **Cluster Assignment and Visualization:**
After fitting the KMeans model, we predict the cluster labels for each customer. Each cluster is plotted using a different color, and the cluster centers are marked with a star symbol to visualize the central points of each cluster.

* **Cluster Labels:**
We assign descriptive labels to each cluster based on customer characteristics. For example, a cluster may represent customers who are high earners and high spenders. These labels are added to the DataFrame for easy interpretation.

* **Final Output:**
The resulting DataFrame includes both the cluster assignment and the descriptive cluster labels, helping to segment customers based on their income and spending behavior.

```python
# Cell 8: Customer Segmentation Analysis Using K-Means

import pandas as pd  # For data manipulation
import matplotlib.pyplot as plt  # For creating plots and visualizations
from sklearn.cluster import KMeans  # For KMeans clustering
from sklearn.preprocessing import MinMaxScaler  # For normalizing the data

# Normalize the 'Annual Income (k$)' column
scaler = MinMaxScaler()  # Initialize the MinMaxScaler to scale data to a range [0, 1]
scaler.fit(df[['Annual Income (k$)']])  # Fit the scaler to the 'Annual Income' data
df['Annual Income (k$)'] = scaler.transform(df[['Annual Income (k$)']])  # Transform and normalize the 'Annual Income'

# Normalize the 'Spending Score (1-100)' column
scaler.fit(df[['Spending Score (1-100)']])  # Fit the scaler to the 'Spending Score' data
df['Spending Score (1-100)'] = scaler.transform(df[['Spending Score (1-100)']])  # Transform and normalize the 'Spending Score'

# Scatter plot of 'Annual Income (k$)' vs 'Spending Score (1-100)' after scaling
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')

# Apply KMeans clustering with 5 clusters
km = KMeans(n_clusters=5)  # Create KMeans model with 5 clusters
y_pred = km.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])  # Fit the model and predict cluster labels

df['cluster'] = y_pred  # Add the predicted cluster labels to the dataframe

# Print the cluster centers
print(km.cluster_centers_)

# Assign data points to specific clusters
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
df4 = df[df.cluster == 3]
df5 = df[df.cluster == 4]

# Plot each cluster with a different color
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color='black')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color='blue')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color='green')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color='red')
plt.scatter(df5['Annual Income (k$)'], df5['Spending Score (1-100)'], color='purple')

# Plot cluster centers with a star marker
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='purple', marker='*')

# Labels and title
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation')

# Show the legend and plot
plt.legend()
plt.show()

# Selecting features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values  # Features for clustering

# Apply KMeans again with the selected features
kmeans = KMeans(n_clusters=5, random_state=42)  # KMeans with 5 clusters
df['cluster'] = kmeans.fit_predict(X)  # Fit and predict the clusters again

# Assigning descriptive labels to each cluster
cluster_labels = {
    0: 'Moderate Earners, Moderate Spenders',
    1: 'High Earners, High Spenders',
    2: 'Low Earners, High Spenders',
    3: 'High Earners, Low Spenders',
    4: 'Low Earners, Low Spenders'
}

# Add a new column with the descriptive labels for each cluster
df['Cluster_Label'] = df['cluster'].map(cluster_labels)

# Display the updated DataFrame
df
```

#Silhouette Score for Cluster Evaluation
The Silhouette Score is a metric used to evaluate the quality of clusters produced by a clustering algorithm, such as K-Means. It measures how similar each point is to its own cluster compared to other clusters.

* **Silhouette Score Calculation:**
The silhouette_score function from `sklearn.metrics` computes the silhouette score based on the cluster assignments. It considers both:

* Cohesion: How close the data points are within the same cluster.
* Separation: How distinct the clusters are from each other.

* **Interpretation of Silhouette Score:**

* A silhouette score close to +1 indicates that the data points are well-clustered, with a high level of cohesion and a large separation between clusters.
* A score near 0 suggests overlapping clusters, meaning that the data points are not clearly assigned to a single cluster.
* A score close to -1 indicates that data points may be assigned to the wrong clusters, leading to poor clustering performance.
* **Purpose:**
The silhouette score helps assess the optimality of the clustering result, guiding decisions on whether the chosen number of clusters (k) is appropriate. Higher values generally suggest a better clustering configuration.


```python
# Cell 9: Silhouette Score Calculation

from sklearn.metrics import silhouette_score  # For calculating the silhouette score

# Calculate the silhouette score
silhouette_avg = silhouette_score(X, df['cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```
